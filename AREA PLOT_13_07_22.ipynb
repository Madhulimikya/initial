```python
import numpy as np
import matplotlib.pyplot as plt
#create data
x=range(1,6)
y=[1,4,6,8,4]
#area plot
plt.fill_between(x,y)
plt.show()
```

```python
x=range(1,15)
y=[1,4,6,8,4,5,3,2,4,1,5,6,8,7]
plt.fill_between(x,y,color='skyblue',alpha=0.75)
plt.show()
```

```python
x=range(1,15)
y=[1,4,6,8,4,5,3,2,4,1,5,6,8,7]
plt.fill_between(x,y,color='skyblue',alpha=0.75)
plt.show()
plt.plot(x, y, color="Slateblue", alpha=0.6)
```
